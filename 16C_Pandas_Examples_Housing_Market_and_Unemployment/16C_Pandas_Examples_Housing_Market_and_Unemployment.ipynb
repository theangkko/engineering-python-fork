# Pandas Examples: Housing Market and Unemployment

### Housing Market

- The [Freddie Mac House Price Index (FMHPI)](https://www.freddiemac.com/research/indices/house-price-index) provides a measure of typical price inflation for houses within the United States since 1975.
- FMAC-HPI_NY.xlsx contains the seasonally adjusted (SA) index of New York State.
- FMAC-HPI_CA_IL_NY_TX.xlsx contains the SA index of four states: California, Illinois, New York, and Texas.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('FMAC-HPI_NY.xlsx', 'Sheet1', index_col = 'Date', na_values='NaN')
df
```

```python
df.plot(figsize=(5,3))
plt.show()
```

```python
df = pd.read_excel('FMAC-HPI_CA_IL_NY_TX.xlsx', 'Sheet1', index_col = 'Date', na_values='NaN')
df
```

```python
df.plot(figsize=(5,3))
plt.legend(loc='best')
plt.show()
```

- December 2000 is the base month in FMHPI. 
- We can move the reference point to the beginning, i.e., January 1975.

```python
new_df = (df - df.iloc[0,:]) / df.iloc[0,:]
new_df.plot(figsize=(5,3))
plt.legend(loc='best')
plt.show()
```

```python
ctable = new_df.corr()  # Correlation table
ctable
```
